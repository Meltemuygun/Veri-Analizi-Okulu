````python
def maxim(a,b,c):
    if a >= b & a >= c:
        return a
    elif b > a & b > c:
        return b
    else:
        return c
````

Ä°pucu 1: `for` dÃ¶ngÃ¼sÃ¼ ve `if` ifadesini birlikte kullanÄ±n.  
Ä°pucu 2: `count()` metodu, bir listede belirli bir elemanÄ±n kaÃ§ kez geÃ§tiÄŸini sayar.

```python
Ã¶rnek_liste = [1,1,3,3,3,3,4,4,4,5,7,7,7]

````python
def benzersiz_yapma(bir_liste):
    return list(set(bir_liste))
````

````python
Ã¶rnek_liste = [1,1,3,3,3,3,4,4,4,5,7,7,7]
print(benzersiz_yapma(Ã¶rnek_liste))
````

````python
def benzersiz(x):
    kopya_liste = x.copy()      # listeyi kopyala (dinamik olarak deÄŸiÅŸtireceÄŸiz)
    benzersiz_liste = []        # sonucu tutacak liste

    for eleman in x:            # x iÃ§indeki her eleman iÃ§in
        if kopya_liste.count(eleman) > 1:   # eleman 1'den fazla kez geÃ§iyorsa
            kopya_liste.remove(eleman)      # bir tanesini sil
        else:                                # sadece 1 kez geÃ§iyorsa
            benzersiz_liste.append(eleman)  # sonuca ekle

    return benzersiz_liste

Ã¶rnek_liste = [1,1,3,3,3,3,4,4,4,5,7,7,7]

print(benzersiz(Ã¶rnek_liste))
````

4- Bir sayÄ± parametresi alan ve sayÄ±nÄ±n asal olup olmadÄ±ÄŸÄ±nÄ± kontrol eden bir Python fonksiyonu yazÄ±n.

Ä°pucu:  
Asal sayÄ±, 1â€™den bÃ¼yÃ¼k ve kendisi ile 1 dÄ±ÅŸÄ±nda pozitif bÃ¶leni olmayan doÄŸal sayÄ±dÄ±r.


````python
def asal_mi(x):
    asal_mi = True

    if type(x) == float or x <= 1:   # x float ise veya 1'den kÃ¼Ã§Ã¼k/eÅŸitse
        asal_mi = "TanÄ±msÄ±z"
    else:
        for i in range(2, x):        # 2'den x-1'e kadar dene
            if x % i == 0:           # tam bÃ¶lÃ¼nÃ¼yorsa
                asal_mi = False
                break

    return asal_mi
````

````python
print(asal_mi(5.3))
````
